```python
import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] =movies['genres'].apply(convert)
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head(1)
```

```python
def convert(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break 
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head(1)
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies["crew"] = movies['crew'].apply(fetch_director)
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[ i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[ i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[ i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[ i.replace(" ","") for i in x])
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df=movies[["movie_id","title","tags"]]
new_df['tags']=new_df['tags'].apply(lambda x : " ".join(x))
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: x.lower())
new_df.head()
```

# Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words= 'english', max_features=5000)
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_stop_words()
```

```python
cv.get_feature_names()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return  " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

# Main function

```python
def recommend(movie):
    # Ensure movie title is found
    matches = new_df[new_df['title'] == movie]
    if matches.empty:
        print(f"Movie '{movie}' not found in the dataset.")
        return
    
    # Get the index of the movie
    movie_index = matches.index[0]
    
    # Assuming similarity is a matrix where similarity[movie_index] gives distances
    distances = similarity[movie_index]
    
    # Sorting distances and getting the top 5 similar movies (excluding the movie itself)
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    # Print recommended movies
    for i in movies_list:
        print(new_df.iloc[i[0]].title)


```

```python
recommend('Spider-Man 2')
```
