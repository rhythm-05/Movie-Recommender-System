```python
import ast
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
movies['crew'] = movies['crew'].apply(fetch_director)
#movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.sample(5)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
#new.head()
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
vector = cv.fit_transform(new['tags']).toarray()
vector.shape
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def filter_movies(df, selected_genres=None, year_range=None):
    filtered = df.copy()
    
    if selected_genres:
        filtered = filtered[filtered['genres'].apply(
            lambda x: any(genre in x for genre in selected_genres)
        )]
    
    if year_range:
        filtered['year'] = pd.to_datetime(filtered['release_date']).dt.year
        filtered = filtered[
            (filtered['year'] >= year_range[0]) & 
            (filtered['year'] <= year_range[1])
        ]
    
    return filtered
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    
    recommendations = []
    for i in distances[1:6]:
        movie_data = new.iloc[i[0]]
        details = fetch_movie_details(movie_data['movie_id'])
        
        recommendations.append({
            'title': movie_data['title'],
            'poster': details['poster'] if details else None,
            'year': movie_data['year'],
            'language': movie_data['original_language'],
            'tmdb_rating': details['rating'] if details else None,
            'imdb_rating': fetch_imdb_rating(movie_data['title'], movie_data['year']),
            'overview': ' '.join(movie_data['tags'].split()[:20]) + '...'  # First 20 words
        })
    
    return recommendations
    
```

```python
import requests

# 🔑 PASTE YOUR API KEY HERE!
TMDB_API_KEY = "68fa86877a10fe6349fff68a23f62007"  # Example: "5a3b6c9d2e1f4g7h8i9j0k"

def get_movie_reviews(movie_name):
    # Step 1: Find the movie ID (like finding a player in a game!)
    search_url = f"https://api.themoviedb.org/3/search/movie?api_key={TMDB_API_KEY}&query={movie_name}"
    search_result = requests.get(search_url).json()
    
    if not search_result.get("results"):
        print("🚫 Oops! Movie not found. Try another name?")
        return []
    
    movie_id = search_result["results"][0]["id"]  # Grab the 1st movie's ID
    
    # Step 2: Get reviews (like unlocking bonus levels!)
    reviews_url = f"https://api.themoviedb.org/3/movie/{movie_id}/reviews?api_key={TMDB_API_KEY}"
    reviews_result = requests.get(reviews_url).json()
    
    if not reviews_result.get("results"):
        print("😢 No reviews yet. Be the first to review?")
        return []
    
    # Step 3: Show top 5 reviews (like high scores!)
    reviews = [review["content"] for review in reviews_result["results"][:5]]
    return reviews

# 🎮 Let's play! Type a movie name:
movie_name = input("🎬 Which movie? (e.g., Inception): ")
reviews = get_movie_reviews(movie_name)

if reviews:
    print("\n🔥 TOP REVIEWS:")
    for i, review in enumerate(reviews, 1):
        print(f"\n⭐ Review {i}: {review}")
else:
    print("❌ Try another movie!")
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

def get_simple_verdict(compound_score):
    if compound_score >= 0.05:
        return "GOOD 😊"
    elif compound_score <= -0.05:
        return "BAD 😠"
    else:
        return "NEUTRAL 😐"

def analyze_movie_sentiment(reviews):
    analyzer = SentimentIntensityAnalyzer()
    total_score = 0
    
    print("\n🔍 REVIEW VERDICTS:")
    for i, review in enumerate(reviews, 1):
        score = analyzer.polarity_scores(review)
        verdict = get_simple_verdict(score['compound'])
        total_score += score['compound']
        
        print(f"\n⭐ Review {i}: {verdict}")
        print(f"   \"{review[:100]}...\"")  # Show first 100 chars
    
    # Overall analysis
    average_score = total_score / len(reviews)
    overall_verdict = get_simple_verdict(average_score)
    
    print("\n🎬 OVERALL MOVIE VERDICT:", overall_verdict)
    print(f"(Average Score: {average_score:.2f})")

# Add this after your existing code where you get the reviews:
if reviews:
    analyze_movie_sentiment(reviews)
else:
    print("❌ No reviews to analyze!")
```
